# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
#Getting the dataset from my files and viewing the data
df = pd.read_excel(r"C:\Cleaning Data\Customer Call List.xlsx")
df
```

#### To start i am going to remove duplicates

```python
#Dropping duplicates
df = df.drop_duplicates()
df
```

```python
#Next I am going to drop columns I do not need
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#To start with I am going to clean the Last_Name column
#As the dataset isn't large I have scanned the column and removed the values that are not needed
df["Last_Name"] = df["Last_Name"].str.strip('123._/')
df
```

```python
#Next we are cleaning the Phone_Number column, strip won't work as there are issues withing the number rather than at the edges
#To fix this I am using the replace function to replace with nothing
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
#Next I am going to format the Phone_Number column to be the same by changing it to a string then then add "-" inbetween certain values
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-'+ x[3:6] + '-' + x[6:10])
df
```

```python
#Finished cleaning the phone number by replacing the sections without phone numbers with blanks
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
#Next I am going to seperate the address into 3 columns["Street_Address", "State", "Post_Code"] so that it is easier to view
#To do this i am doing to use the split function
df[["Street_Address", "State", "Post_Code"]] = df["Address"].str.split(",", n = 2, expand = True)
df
```

```python
#Next I am going to change the values in PayingCustomer to either Y or N corrisponding to what has been entered. This will make it easier to filter
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No","N")
df
```

```python
#I am repeating the same process as PayingCustomer
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N")
df
```

```python
#Replacing all N/a, Nan and Na values with blanks
df = df.replace('Nan','')
df= df.replace('N/a','')
df = df.fillna('')
df
```

```python
#Dropping rows where there has been a do not contact applied as we cannot contact these customers
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] =='Y':
        df.drop(x, inplace = True)
df
```

```python
#Dropping rows where we do not have phone numbers as we cannot call these clients
for x in df.index:
    if df.loc[x,"Phone_Number"] =='':
        df.drop(x, inplace = True)
df
```

```python
#Finally to finish cleaning the data reseting the index uding drop=True as to not create another column.
df = df.reset_index(drop=True)
df
```

```python
df.to_csv(r"C:\Cleaning Data\Customer Call List Completed.csv", index=True)
```
